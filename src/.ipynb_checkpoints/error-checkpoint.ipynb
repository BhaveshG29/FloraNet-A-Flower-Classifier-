```python
import json
import torch as t
from torch.utils.data import Dataset, ConcatDataset, random_split
from torchvision import transforms
from torchvision.transforms import RandAugment
from torchvision.datasets import Flowers102, ImageFolder
from typing import Tuple


#Dataset Wrappers
class Oxford102Wrapper(Dataset):
    def __init__(self, dataset, l_id_to_name, g_name_to_id): #l_id_to_name ==> Local Ids to Name And g_name_to_id ==> Global name to ids
        self.dataset = dataset
        self.l_id_to_name = l_id_to_name
        self.g_name_to_id = g_name_to_id

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        img, l_label = self.dataset[idx]
        class_name = self.l_id_to_name[l_label]
        g_labels = self.g_name_to_id[class_name]

        return img, g_labels


class Oxford17Wrapper(Dataset):
    def __init__(self, dataset, l_id_to_name, g_name_to_id):
        self.dataset = dataset
        self.l_id_to_name = l_id_to_name
        self.g_name_to_id = g_name_to_id

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        img, l_label = self.dataset[idx]
        class_name = self.l_id_to_name[l_label]
        g_label = self.g_name_to_id[class_name]

        return img, g_label

#Wrapper Class to data_transform
class Data_Transform_Wrapper(Dataset):
    def __init__(self, dataset, transform):
        self.dataset = dataset
        self.transform = transform

    def __getitem__(self, idx):
        img, label = self.dataset[idx]
        return self.transform(img), label

    def __len__(self):
        return len(self.dataset)

#Transformations
def data_transform(train: bool = True):
    mean = [0.42575, 0.4012, 0.28805]
    std = [0.23945, 0.20345, 0.19975]

    if train:
        return transforms.Compose([
            #Data Augmentation for Training Datasets aka Better Generalization  
            # transforms.RandomResizedCrop(64, scale=(0.8, 1.0)),
            transforms.Resize((64,64)),
            transforms.RandomHorizontalFlip(),
            RandAugment(num_ops=2, magnitude=6),
            transforms.ToTensor(),
#             transforms.Normalize(mean, std),
        ])

    return transforms.Compose([
        transforms.Resize((128, 128)), 
        transforms.ToTensor(),
#         transforms.Normalize(mean, std),
    ])




#Dataset Builder Function OR Returns Combined Dataset
def FloraNet(
    # class_names: Dict[int, str], 
    train_split: float = 0.8, seed: int = 423) -> Tuple[Dataset, Dataset]:
    

    class_names = {
    0: "alpine sea holly",
    1: "buttercup",
    2: "fire lily",
    3: "anthurium",
    4: "californian poppy",
    5: "foxglove",
    6: "artichoke",
    7: "camellia",
    8: "frangipani",
    9: "azalea",
    10: "canna lily",
    11: "fritillary",
    12: "ball moss",
    13: "canterbury bells",
    14: "garden phlox",
    15: "balloon flower",
    16: "cape flower",
    17: "gaura",
    18: "barbeton daisy",
    19: "carnation",
    20: "gazania",
    21: "bearded iris",
    22: "cautleya spicata",
    23: "geranium",
    24: "bee balm",
    25: "clematis",
    26: "giant white arum lily",
    27: "bird of paradise",
    28: "colt's foot",
    29: "globe thistle",
    30: "bishop of llandaff",
    31: "columbine",
    32: "globe-flower",
    33: "black-eyed susan",
    34: "common dandelion",
    35: "grape hyacinth",
    36: "blackberry lily",
    37: "corn poppy",
    38: "great masterwort",
    39: "blanket flower",
    40: "cyclamen",
    41: "hard-leaved pocket orchid",
    42: "bolero deep blue",
    43: "daffodil",
    44: "hibiscus",
    45: "bougainvillea",
    46: "desert-rose",
    47: "hippeastrum",
    48: "bromelia",
    49: "english marigold",
    50: "japanese anemone",
    51: "king protea",
    52: "peruvian lily",
    53: "stemless gentian",
    54: "lenten rose",
    55: "petunia",
    56: "sunflower",
    57: "lotus",
    58: "pincushion flower",
    59: "sweet pea",
    60: "love in the mist",
    61: "pink primrose",
    62: "sweet william",
    63: "magnolia",
    64: "pink-yellow dahlia",
    65: "sword lily",
    66: "mallow",
    67: "poinsettia",
    68: "thorn apple",
    69: "marigold",
    70: "primula",
    71: "tiger lily",
    72: "mexican aster",
    73: "prince of wales feathers",
    74: "toad lily",
    75: "mexican petunia",
    76: "purple coneflower",
    77: "tree mallow",
    78: "monkshood",
    79: "red ginger",
    80: "tree poppy",
    81: "moon orchid",
    82: "rose",
    83: "trumpet creeper",
    84: "morning glory",
    85: "ruby-lipped cattleya",
    86: "wallflower",
    87: "orange dahlia",
    88: "siam tulip",
    89: "water lily",
    90: "osteospermum",
    91: "silverbush",
    92: "watercress",
    93: "oxeye daisy",
    94: "snapdragon",
    95: "wild pansy",
    96: "passion flower",
    97: "spear thistle",
    98: "windflower",
    99: "pelargonium",
    100: "spring crocus",
    101: "yellow iris",
    102: "bluebell",
    103: "cowslip",
    104: "snowdrop"
}

    #Global label mapping
    sorted_names = class_names.values()
#     sorted_names = sorted(class_names.values()) # Getting Idx of Labels
    g_name_to_id = {name: idx for idx, name in enumerate(sorted_names)} #Reversing the Labels from int:str to str:int

    #Saving Reversed labels to JSON File
    json_path= "../data/labels.json"
    with open(json_path, "w") as f:
        json.dump(g_name_to_id, f, indent=4)

    #Getting Oxford-102 && Dataset-1 = Oxford-102 
    dataset1_train = Flowers102(root="../data", split="train", download=True)
    dataset1_val = Flowers102(root="../data", split="val", download=True)
    dataset1_test = Flowers102(root="../data", split="test", download=True)
        

    dataset1_id_to_name = {k: v for k, v in class_names.items() if k < 102} #Since Flower102 does not come with Labels we create its labels dict.

    dataset1_train = Oxford102Wrapper(dataset1_train, dataset1_id_to_name, g_name_to_id) #Converts it into Pytorch Dataset and makes slicing possible and converts local_labels to Global Labels
    dataset1_val = Oxford102Wrapper(dataset1_val, dataset1_id_to_name, g_name_to_id)
    dataset1_test = Oxford102Wrapper(dataset1_test, dataset1_id_to_name, g_name_to_id)

    #Dataset-2 = Oxford-17
    dataset2 = ImageFolder(root="../data/flowers-17/")  #Oxford-17 Dataset == dataset2     

    # print(len(dataset2)) #Total 1355 Images

    dataset2_name_to_id = dataset2.class_to_idx #Reverse Label's dict of Oxford-17

    dataset2_id_to_name = {idx: name for name, idx in dataset2_name_to_id.items()} #Label's Dictionary of Oxford-17

    dataset2 = Oxford17Wrapper(dataset2, dataset2_id_to_name, g_name_to_id) #Same Function as Oxford102Wrapper

    ##Concatenating Datasets
    #merged_dataset = ConcatDataset([dataset1_train, dataset1_val, dataset2, dataset1_test])

     #Creating Test and Train Split
    test_split = 1 - train_split
    test_len = int(len(dataset2) * test_split)
    train_len = len(dataset2) - test_len
    

    if((test_len + train_len) != len(dataset2)):
        train_len += len(dataset2) - (train_len + test_len) #As int round-offs some images may get left. Causing Incomplete dataset.

    #train_dataset, test_dataset = random_split(merged_dataset, [train_len, test_len], generator=t.Generator().manual_seed(seed)) #Train-Test Split
  

    dataset2_train, dataset2_test = random_split(dataset2, [train_len, test_len], generator=t.Generator().manual_seed(seed))

    train_dataset = ConcatDataset([dataset1_train, dataset1_val, dataset2_train])
    test_dataset = ConcatDataset([dataset1_test, dataset2_test])

    train_dataset, test_dataset = Data_Transform_Wrapper(train_dataset, data_transform(train=False)), Data_Transform_Wrapper(test_dataset, data_transform(train=False))

    return train_dataset, test_dataset
```
